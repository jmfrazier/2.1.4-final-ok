```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from pylab import *
```

```python
df = pd.read_csv(r'C:\Users\jmfra\OneDrive\Documents\Thinkful Data Science Files\2.2.4 challenge data set\LeagueofLegendsEdit.csv')

season_names = df['Season'].unique()
data_s = []
for x in season_names:
    end_season = df[df['Season'] == x]
    final_data = end_season['gamelength']
    data_s.append(final_data)

league_names = df['League'].unique()
data_l = []
for y in league_names:
    end_league = df[df['League'] == y]
    final_data_league = end_league['gamelength']
    data_l.append(final_data_league)
    
xp = df.iloc[0:758,0:8]
red_team_names = xp['redTeamTag'].unique()
data_r = []
for b in red_team_names:
    end_red_team = xp[df['redTeamTag'] == b]
    final_red_team = end_red_team['gamelength']
    data_r.append(final_red_team)

xp = df.iloc[0:758,0:8]
blue_team_names = xp['blueTeamTag'].unique()
data_b = []
for b in blue_team_names:
    end_blue_team = xp[df['blueTeamTag'] == b]
    final_blue_team = end_blue_team['gamelength']
    data_b.append(final_blue_team)    
    

plt.figure(figsize = (17, 5))
plt.boxplot(x=data_s, labels = (season_names))
plt.ylabel('game length in minutes', size = 20)
plt.title('League versus game length')
plt.grid()
gca().set_position((.1, .3, .8, .6))
figtext(.05, .05, 'This graph shows how the meta during each time period affected the average game length. \nIt also can show how the teams play in different styles based on how important the currect season is. \nDuring important international competitions (winter is not a real full season) teams tend to play safer as to not make a mistake')


plt.figure(figsize = (17, 5))
plt.boxplot(x=data_l, labels = (league_names))
plt.ylabel('game length in minutes', size = 20)
plt.title('Season versus game length')
plt.grid()
gca().set_position((.1, .3, .8, .6))
figtext(.05, .05, 'This graph shows how the teams from each region interpret the game differently. \nInstead of showing game style over time, it shows game style in comparison to other regions in general. \nOddly enough, when the teams come together for an international, the range of data actually shrunk and became more normal, meaning the regions adapted to each other instead of playing their own style.')


plt.figure(figsize = (25, 10))
plt.boxplot(x=data_r, labels = (red_team_names))
plt.ylabel('game length in minutes', size = 30)
plt.title('individual team on the red side versus game length', size = 30)
plt.xticks(size = 30)
plt.grid()


plt.figure(figsize = (25, 10))
plt.boxplot(x=data_b, labels = (blue_team_names))
plt.ylabel('game length in minutes', size = 30)
plt.title('individual team on the red side versus game length', size = 30)
plt.xticks(size = 30)
plt.grid()
gca().set_position((.15, .25, .8, .6))
figtext(.1, .1, 'These two graphs together show the difference each specific teams play style based one the sid eof the map they start on. \nBecause the map is not exactly a mirror, teams play different strategies based around neutral objectives and power spikes.', fontsize = 15)
plt.show()
```
